## Password Strenght Checker

## How to Create a Password Strength Checker?
A password strength checker works by understanding the combination of digits, letters, and special symbols you use in your password. It is created by training a machine learning model on a labelled dataset of different combinations of letters and special symbols people use in passwords. The model learns from data about what combinations of letters and symbols can be classified as a solid or weak password.

```python
import pandas as pd
import re
import joblib

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv(
    "Data\PWS_Check.csv",
    sep=",",
    engine="python",
    on_bad_lines="skip"
)
```

```python
df.head()
```

```python
df["strength"] = df["strength;;;;"].astype(str).str.replace(r";.*", "", regex=True)
```

```python
df = df.drop("strength;;;;", axis=1)
```

```python
df.head()
```

```python
df = df.dropna(subset=["password"])
df["password"] = df["password"].astype(str)
```

```python
import math
from collections import Counter

def entropy(password):
    counts = Counter(password)
    length = len(password)
    return -sum((c/length) * math.log2(c/length) for c in counts.values())
```

```python
# ------------------------
# 2. FEATURE ENGINEERING
# ------------------------
def extract_features(password):
    password = str(password)

    return [
        len(password),
        int(bool(re.search(r"[A-Z]", password))),
        int(bool(re.search(r"[a-z]", password))),
        int(bool(re.search(r"[0-9]", password))),
        int(bool(re.search(r"[!@#$%^&*(),.?\":{}|<>]", password))),
        entropy(password)  
    ]
X = df["password"].apply(extract_features).tolist()
y = df["strength"]
```

```python
# ------------------------
# 3. TRAIN
# ------------------------
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)

print("Accuracy:", model.score(X_test, y_test))
```

```python
# ------------------------
# 4. MODEL KAYDET
# ------------------------
joblib.dump(model, "PWS_Model.pkl")

print("Model kaydedildi ✅")
```
