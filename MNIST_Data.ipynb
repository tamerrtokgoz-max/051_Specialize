```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_openml
```

```python
data = fetch_openml("mnist_784", version=1)
print(data)
```

```python
x, y = data["data"], data["target"]
print(x.shape)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
image = np.array(xtrain.iloc[0]).reshape(28, 28)
plt.imshow(image)
```

```python
image = np.array(xtest.iloc[0]).reshape(28, 28)
plt.imshow(image)
```

```python
image = np.array(xtest.iloc[1]).reshape(28, 28)
plt.imshow(image)
```

```python
from sklearn.ensemble import RandomForestClassifier
import joblib

# Modelin eğitilmesi (Hızlı olması için Random Forest seçtik)
model = RandomForestClassifier(n_estimators=100, n_jobs=-1, random_state=42)
model.fit(xtrain, ytrain)

# Başarı oranını kontrol etme
print(f"Model Skoru: {model.score(xtest, ytest)}")

# Modeli kaydetme (Hugging Face'e yüklemek için lazım)
joblib.dump(model, "mnist_model.joblib")
```
